what is Count Plot

count plot are used to when we have a categorical data , variables , and you want to visulize the distribution of these catrgories

```python
sns.countplot(x='class',data=df)
```

```python
sns.countplot(y='source_city',data=df)
```

what is piechart (used for categorial data)

piechart represent the parts of a whole entity.

```python
df['source_city'].value_counts().plot(kind='pie',autopct='%.2f')
```

Donut Plot

above are used to store the show the proportion of each category in relation to thw whole

```python
plt.pie(df['source_city'].value_counts(),
        labels=df['source_city'].unique(),
        autopct='%.2f',wedgeprops=dict(width=0.8))
```
