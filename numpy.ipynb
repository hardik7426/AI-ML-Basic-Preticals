```python
import pandas as pd

# Load dataset
iris = pd.read_csv("Iris.csv")

# Function
def yes_no(SepalLengthCm):
    return 'YES' if SepalLengthCm >= 5 else 'NO'

# Apply function
iris['Result'] = iris['SepalLengthCm'].apply(yes_no)

# Display output
print(iris.head())
```
